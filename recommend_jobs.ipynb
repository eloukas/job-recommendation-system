```python
from libs.helper import *
import os 
import pandas as pd
from sklearn.decomposition import TruncatedSVD 
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
train_dataset_filepath = os.path.join(DATA_FILEPATH, 'vectorized_corpus.csv')
train_dataset = pd.read_csv(train_dataset_filepath)
```

```python
train_dataset.head()
```

```python
train_dataset_to_process = train_dataset.copy()
```

```python
train_dataset_to_process.head()
```

```python
print(COLUMNS_TO_NORMALIZE)
```

```python
for col in COLUMNS_TO_NORMALIZE:
    scaler = MinMaxScaler()
    train_dataset_to_process[col] = scaler.fit_transform(train_dataset_to_process[[col]])
    
```

```python
train_dataset_to_process.head()
```

```python
train_dataset_to_process.drop('job_id', 1, inplace=True)
```

```python
train_dataset_to_process.head()
```

```python
train_dataset_to_process.drop('account_id', 1, inplace=True)
```

```python
train_dataset_to_process.head()
```

```python
train_dataset_to_process.fillna(0.0, inplace=True)
```

```python
train_dataset_to_process
```

```python
similarity_matrix = cosine_similarity(train_dataset_to_process, train_dataset_to_process)
```

```python
print(similarity_matrix)
```

```python
similarity_matrix_df = pd.DataFrame(similarity_matrix)
```

```python
similarity_matrix_df.head()
```

```python
jobs_to_return = 5
def job_recommend(id, similarity_matrix):

    print(f"Input:")
    print(f"Title: {job_to_title[id]}")
    print(f"Description: {job_to_description[id]}\n\n\n")
    
    temp_df = similarity_matrix.iloc[id].nlargest(jobs_to_return+1) # Get top n similar jobs
    list_with_similar_job_ids = list(temp_df.iloc[1:].index) # Discard yourself
    
    # Return similarity scores, titles, descriptions
    similarity_scores = list(temp_df.iloc[1:].values) 
    for id, sim_score  in zip(list_with_similar_job_ids, similarity_scores):
    
        print(f"Cosine Similarity Score: {sim_score}")
        print(f"Title: {job_to_title[id]}")
        print(f"Description: {job_to_description[id]}\n")
```

```python
job_recommend(0, similarity_matrix_df)
```

```python
svd = TruncatedSVD(n_components=100, n_iter=7, random_state=42)
svd.fit(train_dataset_to_process)
```

```python
import pickle
with open('svd.pkl', 'wb') as fout:
    pickle.dump(svd, fout)
```

```python
train_dataset_reduced_dims = pd.DataFrame(svd.transform(train_dataset_to_process))
```

```python
train_dataset_reduced_dims.head()
```

```python
similarity_matrix_reduced_dims = pd.DataFrame(cosine_similarity(train_dataset_reduced_dims, train_dataset_reduced_dims))
```

```python
job_recommend(0, similarity_matrix_reduced_dims)
```
